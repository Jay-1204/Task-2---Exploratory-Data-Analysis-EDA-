```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

### Step:-2 Create Histograms and Boxplots for Numeric Features

```python
# Histogram
df.hist(figsize=(12, 10), bins=30, color='skyblue')
plt.suptitle("Histograms of Numeric Features")
plt.tight_layout()
plt.show()
```

```python
# Boxplots
numeric_cols = df.select_dtypes(include='number').columns
for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f"Boxplot of {col}")
    plt.show()
```

### Step:-3  Use Pairplot / Correlation Matrix for Feature Relationships

```python
numeric_df = df[['Age', 'Fare', 'SibSp', 'Parch','Pclass']]
corr_matrix = numeric_df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Numeric Features")
plt.tight_layout()
plt.show()
```

```python
numeric_df = df[['Age', 'Fare', 'SibSp', 'Parch',]]
sns.pairplot(numeric_df)
plt.suptitle("Pairplot of Numeric Features", y=1.02)
plt.show()
```

### Step:- 4 Identify Patterns, Trends, or Anomalies

```python
# Survival by passenger class
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()

```

```python
# Survival by gender
plt.figure(figsize=(8, 5))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()

```

```python
# Family size analysis
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
plt.figure(figsize=(10, 5))
sns.barplot(x='FamilySize', y='Survived', data=df)
plt.title('Survival Rate by Family Size')
plt.show()
```

```python
# Survival by embarkation port
plt.figure(figsize=(8, 5))
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title('Survival Rate by Embarkation Port')
plt.show()
```

### Step:-5 Feature-Level Inferences

```python
print("\nKey Statistics:")
print(f"Overall survival rate: {df['Survived'].mean()*100:.2f}%")
print("\nSurvival rate by class:")
print(df.groupby('Pclass')['Survived'].mean())
print("\nSurvival rate by gender:")
print(df.groupby('Sex')['Survived'].mean())
print("\nAverage fare by class:")
print(df.groupby('Pclass')['Fare'].mean())
print("\nAverage age by survival:")
print(df.groupby('Survived')['Age'].mean())

# Age groups analysis
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 30, 50, 100], 
                      labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])
print("\nSurvival rate by age group:")
print(df.groupby('AgeGroup')['Survived'].mean())
```
